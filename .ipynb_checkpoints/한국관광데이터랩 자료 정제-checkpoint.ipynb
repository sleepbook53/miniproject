```python
df_acc = df_acc[(df_acc["중분류 카테고리"] != '쇼핑') & \
               (df_acc["중분류 카테고리"] != '음식')]
df_acc
```

```python
#정렬하기
df_Gu = df_acc.sort_values(by= '구군명').reset_index(drop = True)
df_Gu
```

```python
df_Gu_list = df_Gu["구군명"].unique()
df_Gu_list
```

```python
#구별 분리
dic_list = {}
for i in range(0, len(df_Gu_list)) :
    df2 = pd.DataFrame() #깨끗한 데이터 프레임
    df2 = df_Gu[df_Gu["구군명"]== df_Gu_list[i]] #꺠끗한 데이터 프레임에 자료 삽입
    dic_list[df_Gu_list[i]] = df2 #키 값별 데이터 프레임
```
