### 도로명주소 위도 경도 값으로 바꿔주기

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
# 위도, 경도 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y
```

## 에러 검출 및 에러인 요소 전부 '구군 좌표값으로 변환

```python
latitude = []
longitude =[]
for i in range(0, len(address)): #len(address)
    try :
        d = geocoding(address[i])
        print(i, ":번째 / ", d[0], " / ", d[1])
        latitude.append(d[0])
        longitude.append(d[1])

    except:
        # 오류 -> 리스트처리 빈것...
        print(i, '번째 처리 중 예외 발생')
        latitude.append('error')
        longitude.append('error')
```

### Dataframe만들기

```python

address_df = pd.DataFrame({'관광지명': csv['관광지명'],'도로명주소':csv['도로명주소'],\
                           '중분류 카테고리':csv['중분류 카테고리'], '소분류 카테고리':csv['소분류 카테고리'],\
                           '구군명':csv['구군명(영어)'], '주소':address, '위도':latitude,'경도':longitude})

address_df
```

```python
#에러값 확인
address_df[address_df["위도"] == 'error']
```

### 위경도가 error인 요소에 대해 고정값으로 인출

```python
for i in range(0, len(address_df)):
    
    # 위도가 에러가 아니면 넘어감
    if address_df.loc[i, '위도'] != 'error' :
        pass
    
    # 위도가 에러면 아래 코드 실행
    else:
        # 주소를 변경함
        # 주소를 공백 단위로 쪼개어 부산 + 군/구 로 변경
        address_df.loc[i, '주소'] = address_df.loc[i, '주소'].split()[0] + ' ' + address_df.loc[i, '주소'].split()[1]
        # 변경된 주소의 위도 경도 추출하여 삽입
        address_df.loc[i, '위도'] = geocoding(address_df.loc[i, '주소'])[0]
        address_df.loc[i, '경도'] = geocoding(address_df.loc[i, '주소'])[1]

address_df
```
