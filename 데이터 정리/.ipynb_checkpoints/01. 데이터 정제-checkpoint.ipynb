# 관광 데이터랩 자료 정제

```python
#데이터 업로드
df = pd.read_excel("./data/culdata.xlsx")
```

```python
df
```

### 활용할 데이터만 정제

```python
#구군명 생성
df_Gu = df
df_Gu["구군명"]= df["도로명주소"].str.split(" ", expand = True)[1]
```

```python
#활용할 데이터만 정제
df1 = df_Gu[(df_Gu["중분류 카테고리"]== '공연/행사')  |\
    (df_Gu["중분류 카테고리"]== '역사관광지') | \
    (df_Gu["중분류 카테고리"] == '휴양관광지') | \
    (df_Gu["중분류 카테고리"] == '문화시설')].reset_index(drop = True).drop(["순위"], axis =1)

df1
```

```python
# 활용할 데이터 정제 2
df2 = df1.drop(["검색건수"], axis = 1)
df2
```

# 데이터 구별 소계

```python
gu_count = df2.groupby("구군명").count()
gu_count["소계"] = gu_count["소분류 카테고리"]
gu_count = gu_count.drop(["관광지명", "도로명주소", "중분류 카테고리", "소분류 카테고리"], axis = 1)
```

```python
gu_count
```

#### 인구 데이터 소계

```python
df_pop = pd.read_csv("./save/data_pop.csv").drop(["Unnamed: 0"], axis = 1)
df_pop["구군명"] = df_pop["소재지(시군구)"]
df_pop.drop(["소재지(시군구)"], axis = 1)
df_pop = df_pop[['구군명', '전체인구(A) (명)']]
df_pop
```

```python
#한글화
ddd=pd.read_csv("./data/ddd.csv")
gu_pop= ddd.sort_values(by = '구군명', ascending = True)
gu_pop
```

## 구군별 문화시설/인구 비율

```python
#두 데이터 합치기(inner join)
df_r =pd.merge(gu_count, gu_pop, how = 'inner', on = '구군명')
df_r
```

```python
#비율 구하기
df_r["인구 대비 문화시설비"] = df_r["소계"]/df_r["인구(명)"]
```

```python
df_tmp = df2
df_tmp2 = df_r2[['구군명','인구(천명)']]
df_tmp3 =pd.merge(df2, df_tmp2, how = 'inner', on = '구군명') 
df_tmp3
```

```python
df_tmp = df2.groupby(['구군명', '소분류 카테고리']).count()
df_tmp
```

```python
#pd.set_option('display.max_row', 40) - 화면에 프레임 갯수 조절
df_tmp2 = pd.DataFrame()
df_tmp2["집계"] = df_tmp["관광지명"]
df_tmp2
```

### Groupby를 데이터 프레임으로 변경하기

```python
#Groupby를 데이터 프레임으로 변경하기
#.add_suffix('')안의 공백은 count값의 컬럼 명에 추가로 
df_co = df_tmp2.add_suffix('').reset_index()
df_co
```

## 구군별 문화시설별 박스플롯 비교 및 출력

```python
gus = df_co['구군명'].unique()
#cnt = 0
for gu in gus:
    
    # 그래프 사이즈 설정df_co
    fig = plt.figure(figsize = (20, 5))
    
    # x축 y축에 넣을 컬럼 설정 / 기반 데이터 설정
    sns.barplot(x = '소분류 카테고리', y = '집계', data = df_co[df_co['구군명'] == gu])
    
    # 그래프 제목 설정
    plt.title(str(gu) + ' ' + '문화시설 분포')
    
    # x, y축 이름 설정
    plt.ylabel('시설 수')
    plt.xlabel('')
    
    # 출력
    plt.show()
    #cnt  = cnt + 1
   # plt.savefig('savefig_default_{}.png'.format(cnt))
```

```python
# 통계치 분석
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#상관관계분석

df_corr = pd.DataFrame({"문화시설":df_r["소계"], "부동산":df_Bamp['100m'], \
                        '지역별 총 문화소비' :df_moon["총액/6개월,만원"], \
                        '지역별 인당 문화소비' : df_moon2["인당 평균, 원"], \
                        '경제활동인구' : df_w2['경제활동인구 (천명)'],'경제활동참가율' : df_w2['경제활동참가율 (％)'], \
                        '면적' : df_map1["면적"] })

#두집단간 상관관계
corr = df_corr.corr(method = 'pearson')
corr
```

피어슨의 상관계수는 일반적으로,

  값이 -1.0 ~ -0.7 이면, 강한 음적 상관관계

  값이 -0.7 ~ -0.3 이면, 뚜렷한 음적 상관관계

  값이 -0.3 ~ -0.1 이면, 약한 음적 상관관계

  값이 -0.1 ~ +0.1 이면, 없다고 할 수 있는 상관관계

  값이 +0.1 ~ +0.3 이면, 약한 양적 상관관계

  값이 +0.3 ~ +0.7 이면, 뚜렷한 양적 상관관계

  값이 +0.7 ~ +1.0 이면, 강한 양적 상관관계

로 해석됩니다.
