```python
#import
import pandas as pd
import numpy as np

df = pd.read_excel("./data/culdata.xlsx")
```

```python
#검색건수별로 중분류 카테고리 분석
df_midle = df.groupby(["중분류 카테고리"]).sum()
df_midle=df_midle.drop(["순위"], axis = 1)
df_midle.sort_values(by='검색건수', ascending = False)

df_midle
```

```python
#소분류 카테고리별 분류
df_cate = df.groupby(["중분류 카테고리", "소분류 카테고리"]).sum()
df_cate= df_cate.drop(["순위"], axis = 1)
df_cate
```
